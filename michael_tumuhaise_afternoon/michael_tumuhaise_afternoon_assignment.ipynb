Number 5 Write a Pandas program to display the most frequent value in the given series and replace everything else as ‘Other’ in the series. (Use the ‘HomeTeamGoals’ column in the dataset provided)

```python
# to use the HomeTeamGoals’ column we must first read the dataset given to us
# reading the dataset
df = pd.read_csv('AfricaCupofNationsMatches.csv')
df
```

```python
# Calculate the most frequent value we use the mode() function with [0] index set at 0 so that it call run through all the values in the column 
most_frequent_value = df['HomeTeamGoals'].mode()[0]

# Replace everything else as 'Other'
df['HomeTeamGoals'] = df['HomeTeamGoals'].apply(lambda x: most_frequent_value
                                                if x == most_frequent_value 
                                                else 'Other')

# Print the modified series and it will only display the values of that column
print(df['HomeTeamGoals'])
```
